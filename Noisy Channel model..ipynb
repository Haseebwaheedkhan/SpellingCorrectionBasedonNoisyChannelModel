```python
def add_tags(data,n):
    # Take list of sentence as an input
    # At n = 1 unique tags will be added and retrn string of sentence
    # At n = 2 unique tags will be replaced with standard tags i.e <S> and </S>. Finally return list words 
    
    if n == 1:
        all_lines = ""
        for i in data:
            w = "#" # Start of Sentence tag
            for word in i:
                if "\n" in word:
                    w+='&' # End of sentence tag
                # Removing spaces in sentence
                elif " " in word:
                 pass
                else:
                    w+=word
            all_lines = all_lines + w

    elif n == 2:
        all_lines = ""
        for i in data:
            w = "$ " # Start of Sentence tag
            for word in i:
                if "\n" in word:
                    w+=' & ' # End of sentence tag
                else:
                    w+=word
            all_lines = all_lines + w

    elif n == 3:
        all_lines = []
        for i in data:
            if i == "#":
                all_lines.append("#") # Replacing Start of sentence tag with standard tag
            elif i == '&':
                all_lines.append("&") # Replacing End of sentence tag with standard tag
            else:
                all_lines.append(i)
        
        
    return all_lines      
```

```python
def read_file(file_name):
    all_lines = []
    f = open(file_name,'r')
    for w in f:
        all_lines.append(w)
        
    # Print status of file 
    if f:
        print("**** File read successfully ****")
    else:
        print("**** File read unsuccessfully ****")
        print("length of file: ",len(all_lines))
    return all_lines



def tokenize_word(all_lines):
    
    # Input list of words
    # Return words in the form of token
    
    all_lines = add_tags(all_lines,2) #Adding Tags at at begning and end of sentence
    all_lines = all_lines[0:1000] # Taking first 500 sentences for test purpose only
    all_words = word_tokenize(all_lines) #Tokenize words 
    all_words = add_tags(all_words,3)
    return all_words
      
    
file_name = "data.txt"   
all_lines = read_file(file_name)
train_data_tokenize_string = add_tags(all_lines,1)
train_data_tokenize_string = train_data_tokenize_string[0:1000]
train_data_tokenize_words = tokenize_word(all_lines)
```

```python
def ngram_model(words_list, n):
    # Input = List of words
    # n = length of gram e.g n = 2 mean bi-gram model
    # Returns ngram list
    ngram_word = []

    
    for index in range(len(words_list) + 1 - n):
        words = []
        for i in range(n):
            words.append(words_list[index + i])

        ngram_word.append(words)
    return ngram_word
        
train_unigram_words = ngram_model(train_data_tokenize_words, 1) 
train_bigram_char   = ngram_model(train_data_tokenize_string, 2)  
train_unigram_char   = ngram_model(train_data_tokenize_string, 1)  


print(len(train_unigram_words),train_unigram_words)
print(len(train_bigram_char),train_bigram_char)
print(len(train_unigram_char),train_unigram_char)
```

```python
def check_delete(correct_word, misspellword):
    '''
    Compare each character of misspellword with correct_word and identify the character after which new character is to be added
    Arguments:+
    Both arguments of this operator is string
    Return:
    Idetified character i.e string after with new word is to be inserted 
    '''
    inserted_char = []
    insert_status = 0

    for index in range(len(correct_word)-1):

        if len(misspellword)+1 == len(correct_word): # Test if character insert required
            # If mismatch character is found
            if ((correct_word[index] != misspellword[index]) and (index != 0)):
                # Check next character of correct words incase matches with current character of 
                if correct_word[index+1] == misspellword[index]:
                    return ("1", correct_word[index-1], correct_word[index])
            # If first letter of misspell word doesnot macth 
            elif ((correct_word[index] != misspellword[index]) and (index != 1)):
                return ("1", '#', correct_word[index])
            # All char in misspell word matches with correct word  
            elif ((index+1) == len(correct_word)-1):
                return ("1", correct_word[index], correct_word[index+1])
        else:
            return ("0","None","None")
    return (0, "None","None")

def check_insert(correct_word, misspellword):
    '''
    Compare each character of misspellword with correct_word and identify the 
    character to be deleted from misspllword to make it similar to the correctword
    Arguments:
    Both arguments of this operator is string
    Return:
    Idetified characters i.e tuple i.e ( Status i.e 0 or 1, Previous character after deleted, char to be deleted )
    '''
    # 4 5 kisi usaye
    delete_char = ""
    step = 0 # No of deletion performed
    
    if len(correct_word)+1 == len(misspellword):
        for index in range(len(correct_word)):
            if  (correct_word[index] != misspellword[index]):
                if index == 0:
                    step += 1
                    delete_char+=str(step) + correct_word[index] + "#"
                elif (index >=1):
                    if correct_word[index::] == misspellword[index+1::]:
                        step += 1
                        delete_char+=str(step) + correct_word[index] + misspellword[index]
                    else:
                        return ("0", "None", "None")

        if step == 0:
            step += 1
             # Check if all char in correct word is same as of misspellword and only last char is to be removed
            delete_char += str(step) + correct_word[-1] + misspellword[-1]
    #if delete operation is not performed
    else:
        return ("0", "None", "None")
    return (delete_char[0],delete_char[1],delete_char[2])

def check_substite(correct_word,misspell_word):
    '''
    Compare word characterwise and identify the character to be substitued 
    from correct word to make it similar to the correctword
    Arguments:
    Both arguments of this operator is string
    Return:
    Idetified characters i.e tuple i.e (Delete Status i.e 0 or 1, Previous character after deleted, char to be deleted )
    '''
    if (len(misspell_word) == len(correct_word) and (correct_word != misspell_word)):
        for index in range(len(correct_word)):
            # Find mismatch character
            if correct_word[index] != misspell_word[index]:
                # If first chara is a mismatch and rest of the chars in both the words are same
                if ((index == 0) and (correct_word[1:] == misspell_word[1:] )):
                    return ("1", correct_word[index], misspell_word[index])
                # If last char is a mismatch and rest of the chars in both the words are same
                elif ((index == len(correct_word)-1) and (correct_word[0:len(correct_word)-1] == misspell_word[0:len(correct_word)-1] )):
                    return ("1", correct_word[index], misspell_word[index])
                # If char in a word is a mismatch and rest of the chars in both the words are same
                elif ((correct_word[0:index-1] == misspell_word[0:index-1]) and (correct_word[index+1:] == misspell_word[index+1::])):
                    return ("1", correct_word[index], misspell_word[index])
                else:
                    return (0, "None","None")
    else: 
        return (0, "None", "None")

def check_transpose(correct_word,misspell_word):
    '''
    Compare word characterwise and identify the character to be substitued 
    from correct word to make it similar to the correctword
    Arguments:
    Both arguments of this operator is string
    Return:
    Idetified characters i.e tuple i.e (Delete Status i.e 0 or 1, Previous character after deleted, char to be deleted )
    '''
    trans_char = ""
    if (len(misspell_word) == len(correct_word) and (correct_word != misspell_word)):
        for index in range(len(correct_word)):
            # Find mismatch character
            if correct_word[index] != misspell_word[index]:
                if (correct_word[index] == misspell_word[index+1]) and (correct_word[index+1] == misspell_word[index]):
                    if ((len(correct_word) > index + 1) and correct_word[index+2::] == misspell_word[index+2::] ):
                        return ("1", correct_word[index], misspell_word[index])
                else:
                    return ("0", "None", "None")
            # Before the last loop check last element in case mismatch so no transposition possible
            elif ((correct_word[index+1] != misspell_word[index+1]) and (index+1 == len(correct_word)-1)):
                return ("0", "None", "None")
    else:
        return ("0", "None", "None")


```

```python
# Creating Confusion Matrix
def CreateConfusionMatrix():
    alphabets = "abcdefghijklmnopqrstuvwxyz#"
    del_matrix = {}
    ins_matrix = {}
    sub_matrix = {}
    tra_matrix = {}

    for letter in alphabets:
        submatrix = {}
        for l in alphabets:
            submatrix[l] = 0
        del_matrix[letter] = submatrix.copy()
        ins_matrix[letter] = submatrix.copy()
        sub_matrix[letter] = submatrix.copy()
        tra_matrix[letter] = submatrix.copy()
    return (del_matrix, ins_matrix, sub_matrix, tra_matrix) 

(del_matrix, ins_matrix, sub_matrix, tra_matrix)  = CreateConfusionMatrix()
(del_matrix, ins_matrix, sub_matrix, tra_matrix)
```

```python
def PopulateConfusionMartix(misspellword):
    #Creating Confusion Matrix
    (del_matrix, ins_matrix, sub_matrix, tra_matrix)  = CreateConfusionMatrix()
    # misspellword = {"usay" : ['usaye']}
    for correct_word in misspellword:
        for error_word in misspellword[correct_word]:

            # check delete operation 
            if check_delete(correct_word,error_word)[0] == "1":
                result = check_delete(correct_word,error_word)
                del_matrix[result[1]][result[2]] += 1
            # Check insert operation
            elif check_insert(correct_word,error_word)[0] == "1":
                result = check_insert(correct_word,error_word)
                ins_matrix[result[1]][result[2]] += 1
            # Check Substitue Operation       
            elif check_substite(correct_word,error_word)[0] == "1":
                result = check_substite(correct_word,error_word)
                sub_matrix[result[1]][result[2]] += 1
            # Check Transpose operational
            elif check_transpose(correct_word,error_word)[0] == "1":
                result = check_transpose(correct_word,error_word)
                tra_matrix[result[1]][result[2]] += 1
            else :
                pass                 
    return (del_matrix, ins_matrix, sub_matrix, tra_matrix)

(del_matrix, ins_matrix, sub_matrix, tra_matrix) = PopulateConfusionMartix(misspellword)
(del_matrix, ins_matrix, sub_matrix, tra_matrix)
```

```python
def GetOperation (correct_word,error_word):

    if check_delete(correct_word,error_word)[0] == "1":
        oper = check_delete(correct_word,error_word)
        return("DELETE",1,oper[1],oper[2])
    # Check insert operation
    elif check_insert(correct_word,error_word)[0] == "1":
        oper = check_insert(correct_word,error_word)
        return("INSERT",1,oper[1],oper[2])
    # Check Substitue Operation       
    elif check_substite(correct_word,error_word)[0] == "1":
        oper = check_substite(correct_word,error_word)
        return("SUBSTITUE",2,oper[1],oper[2])
    # Check Transpose operational
    elif check_transpose(correct_word,error_word)[0] == "1":
        oper = check_transpose(correct_word,error_word)
        return("TRANSPOSE",2,oper[1],oper[2])
    else :
        return(0, "#," "#")  

print("'usay','usaye' ",GetOperation('usay','usaye'),"\n'usay','usab' ",GetOperation('usay','usab'))
print("'usay','usya' ",GetOperation('usay','usya'),"\n'usay','usy' ",GetOperation('usay','usy'))
```

```python
def ChannelModelProb(correct_word, error_word, misspellword, char_x,char_xy):
    # Find the operation to be performed
    oper = GetOperation(correct_word, error_word) 
    # Get Populated Confusion matrix to find the numenator of Pr(tlc) e.g ins[c,p] 
    (del_matrix, ins_matrix, sub_matrix, tra_matrix) = PopulateConfusionMartix(misspellword)

    if oper[0] == 'DELETE':
        print(oper[0], end = " ")
        num = del_matrix[oper[2]][oper[3]]
        den = char_xy[oper[2],oper[3]]
        print(oper[2],oper[3],num,"/",den)
        return (oper[1], num/den)

    elif oper[0] == 'INSERT':
        print(oper[0], end = " ") 
        num = ins_matrix[oper[2]][oper[3]]
        den = char_x[oper[2]]
        print(oper[2],oper[3],num,"/",den)
        return (oper[1], num/den)
        
    elif oper[0] == 'SUBSTITUE':
        print(oper[0], end = " ")
        num = sub_matrix[oper[2]][oper[3]]
        den = char_x[oper[3]]
        print(oper[2],oper[3],num,"/",den)
        return (oper[1], num/den)

    elif oper[0] == 'TRANSPOSE':
        print(oper[0], end = " ") 
        num = ins_matrix[oper[2]][oper[3]]
        den = char_xy[oper[2],oper[3]]
        print(oper[2],oper[3],num,"/",den)
        return (oper[1], num/den)

print(ChannelModelProb("usay", "usaye", misspellword, char_x,char_xy))
print("\n")
print(ChannelModelProb("usay", "uasy", misspellword, char_x,char_xy))
print("\n")
print(ChannelModelProb("uasy", "usay", misspellword, char_x,char_xy))
```
